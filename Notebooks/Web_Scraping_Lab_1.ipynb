```python
#Function to export the webscrap data to csv and excel.
def save_to_csv(df, filename=None):
    """
    Save the DataFrame to a CSV file
    """
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    if filename is None:
        filename = f"WebScrap_{timestamp}.csv"

    try:
        filename = f"{filename}_{timestamp}.csv"
        df.to_csv(filename, index=False, encoding='utf-8')
        print(f"Data saved to {filename}")
        return True
    except Exception as e:
        print(f"Error saving to CSV: {e}")
        return False

def save_to_excel(df, filename=None):
    """
    Save the DataFrame to an Excel file
    """
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    if filename is None:
        filename = f"WebScrap_{timestamp}.xlsx"

    try:
        filename = f"{filename}_{timestamp}.xlsx"
        df.to_excel(filename, index=False)
        print(f"Data saved to {filename}")
        return True
    except Exception as e:
        print(f"Error saving to Excel: {e}")
        return False
```

```python
#Importing libraries
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import matplotlib.pyplot as plt
import time
import csv
from datetime import datetime
```

```python
#Link to the website
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
```

```python
#To check if the link is reachable
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0 Safari/537.36"}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    print("Request was successful")
else:
    print(f"Request failed, Status code: {response.status_code}")
```

```python
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
soup = bs(response.text, "lxml")
table = soup.find("table", {"id": "constituents"})
```

```python
# Extract headers
headers_row = [th.text.strip() for th in table.find("tr").find_all("th")]

# Extract rows
rows = []
for tr in table.find_all("tr")[1:]:
    cells = [td.text.strip() for td in tr.find_all(["td", "th"])]
    if cells:
        rows.append(cells)

# Convert to DataFrame
df = pd.DataFrame(rows, columns=headers_row)
df.head()
```

```python
save_to_csv(df, "List of S&P 500 companies")
save_to_excel(df, "List of S&P 500 companies")
```
